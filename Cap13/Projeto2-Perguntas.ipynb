## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
cidade_maior_valor_venda = df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda'].sum().idxmax()
valor_maior_valor_venda = df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda'].sum().max()

print("%s, %f" % (cidade_maior_valor_venda, valor_maior_valor_venda))
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
vendas_data = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
print(vendas_data)

plt.bar(vendas_data.index, vendas_data.values)
plt.xlabel('Data')
plt.ylabel('Valor de total de venda')
plt.title('Total de Vendas por Data')
plt.tight_layout()
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
vendas_estado = df_dsa.groupby('Estado')['Valor_Venda'].sum()

plt.bar(vendas_estado.index, vendas_estado.values)
plt.xlabel('Estado')
plt.ylabel('Valor de total de venda')
plt.title('Total de Vendas por Estado')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
vendas_top10_estado = df_dsa.groupby('Estado')['Valor_Venda'].sum().nlargest(10)

plt.bar(vendas_top10_estado.index, vendas_top10_estado.values)
plt.xlabel('Estado')
plt.ylabel('Valor de total de Vendas')
plt.title('Total de Vendas por Estado (TOP10)')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
vendas_segmento = df_dsa.groupby('Segmento')['Valor_Venda'].sum()
segmento_maior_valor = vendas_segmento.idxmax()

explode = [0.1 if segmento == segmento_maior_valor else 0 for segmento in vendas_segmento.index]

plt.pie(vendas_segmento, explode=explode, labels=vendas_segmento.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Total de vendas por segmento')
plt.show()
```
